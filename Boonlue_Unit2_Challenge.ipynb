```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df=pd.read_csv('NOAA_Yearly.csv')
print(df)
```

```python
#Graph 1: Displays the histogram of the annual minimum, average, maximum temperature
#in Fahrenheit from 1990 to 2018 in all weather stations in Santa Clara County, California. 
#The histogram allows for easy visualization of the frequency of the 
#temperatures during this time period. According to Graph 1, the most frequent 
#annual minimum temperature is around 48 degrees (represented by orange).
#The most frequent annual average temperature is 60 degrees (represented by light blue). 
#And the most frequent maximum temperature is 70 degrees (represented by pink).

x=df['TAVG']
y=df['TMAX']
z=df['TMIN']

plt.hist(x,color='deepskyblue',alpha=0.7)
plt.hist(y,color='orchid',alpha=0.7)
plt.hist(z,color='salmon',alpha=0.7)
plt.title('Annual Mininium, Average, and Maximum Temperature from 1990 to 2018')
plt.xlabel('Temperature in Fahrenheit')
plt.ylabel('Frequency')
plt.show()
```

```python
#Graph 2 is a subplot. Graph 2a displays the number of days in a year with maximum temperature 
#above 90 degrees at all weather stations in Santa Clara County between 1990-2018. Graph 2b shows the 
#number of days in a year with minimum temperature under 32 degrees at all stations in the County between
#the same time frame. These visulizations can help potentially identify the impacts of climate change in 
#the San Francisco Bay Area. In Graph 2a, the outlier of 70 days and 80 days around 1997-1998 aligns with the 
#El Nino event (warm phase) that California experienced. Trends in Graph 2a are not well defined due to the 
#difference in the number stations over the years, there are more weather stations in the present than in 1990. 
#The trend is Graph 2b is easier to identify,the number of days with minimum temperature uder 32 degrees have 
#been decreasing despite the greater number of stations in 2018 than previously. This can signify a warming
#trend in the Bay Area over the past 30 years. 

plt.figure(figsize=(20,5))

plt.subplot(1,2,1)
plt.scatter(x=df['DATE'], y=df['DX90'], color='mediumpurple', marker='d')
plt.title('Number of Days with Maximum Temperature over 90 Degrees in Santa Clara County by Year')
plt.xlabel('Year')
plt.xlim([1988,2020])
plt.ylabel('Number of Days')

plt.subplot(1,2,2)
plt.scatter(x=df['DATE'], y=df['DT32'], color='goldenrod', marker='d')
plt.title('Number of Days with Minimum Temperature under 32 Degrees in Santa Clara County by Year')
plt.xlabel('Year')
plt.xlim([1988,2020])
plt.ylabel('Number of Days')
plt.show()
```

```python
df_moffett=pd.read_csv('MoffettField_NOAA.csv')
print(df_moffett)
```

```python
#Graph 3 displays the annual average wind speed at the Moffett Field Weather Station in Santa Clara County, California
#between the year 1990 to 2018. In 1993, the weather station experienced its slowest annual average wind speed of 
#about 3.8 miles per hour. In 2001, the station experienced the highest annual average wind speed of 6.0 miles per hour. 
#There seems to be a decline in wind speed from 2017 to 2018. Because the scientific evidence on correlation between
#annual average wind speed and climate change is vague, no definite conclusions can be made whether the trend observed
#is due to climate change.


x=df_moffett['DATE']
y=df_moffett['AWND']

plt.plot(x,y, color='mediumseagreen')
plt.title('Annual Average Wind Speed at Moffett Field Weather Station from 1990-2018')
plt.xlim([1988,2020])
plt.xlabel('Year')
plt.ylim([3.5,6.3])
plt.ylabel('Wind Speed (miles per hour)')
plt.show()
```

```python
#Graph 4 shows the annual average maximum (crimson) and minimum (blue) temperature between 1990 to 2018. 
#There seems to be an increasing trend in the average annual maximum temperature and a decreasing trend
#in average annual minimum temperature. Highest max. and min. temperatures were observed back in 1992-1993,
#which was an El Nino (warming) event for California. 

plt.scatter(
    x=df_moffett['DATE'],
    y=df_moffett['TMAX'],
    color='crimson',
    marker='X')

plt.scatter(
    x=df_moffett['DATE'],
    y=df_moffett['TMIN'],
    color='steelblue',
    marker='X')

plt.title('Annual Avg. Maximum and Minimum Temperature at Moffett Field 1990-2018')
plt.xlabel('Year')
plt.ylabel('Degrees in Fahrenheit')

plt.show()
```

```python
#Graph 5 shows the box plot for the average annual wind speed for the Moffett Field
#station. There are 2 possible outliers that are below the whiskers.


x=df_moffett.AWND
plt.boxplot(x)
plt.show()
```
